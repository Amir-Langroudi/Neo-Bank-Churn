```python
df_users[df_users["city_clean"].str.contains(r"\d", na=False)]["city_clean"].value_counts().head(20)
```

```python
df_users["city_clean"] = (
    df_users["city"]
    .astype(str)
    .str.strip()
    .str.title()
    .str.replace(r"^\d+\s*", "", regex=True)  # Remove numbers at the beginning
    .str.replace(r"\s+\d+$", "", regex=True)  # Removes numbers at the end
)
```

```python
import re

def clean_city(city):
    if pd.isna(city):
        return city

    city = str(city)

    # remove postal codes like 61-058 or -300
    city = re.sub(r'\b\d{2}-\d{3}\b', '', city)
    city = re.sub(r'^-?\d+\s*', '', city)

    return city.strip()

df_users['city_clean'] = df_users['city_clean'].apply(clean_city)
```

```python
df_users['city_clean'] = df_users['city_clean'].str.replace(r'\s\d+[A-Za-z]*$', '', regex=True)
```

```python
df_users['city_clean'] = (
    df_users['city_clean']
    .str.strip()
    .str.title()
)
```

```python
df_users['city_clean'] = (
    df_users['city_clean']
    .str.replace(r'\d+', '', regex=True)
    .str.replace(r'\s+', ' ', regex=True) # clean extra spaces
    .str.strip()
)
```

```python
import numpy as np

# 1) Clean whitespace and handle string "nan" values
df_users["city_clean"] = (
    df_users["city_clean"]
    .astype(str)              # ensure column is string type
    .str.strip()              # remove leading/trailing spaces
    .replace(["", "nan", "Nan", "None", "NaN"], np.nan)  # replace empty or invalid strings with NaN
)

# 2) Replace missing values with "Unknown"
df_users["city_clean"] = df_users["city_clean"].fillna("Unknown")

# 3) Replace very short city names (<= 2 characters) with "Unknown"
df_users.loc[df_users["city_clean"].str.len() <= 2, "city_clean"] = "Unknown"
```
